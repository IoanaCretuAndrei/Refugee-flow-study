```python
# Calcula el umbral para más del 10% de NaN
threshold = len(surface_temperature.columns) * 0.1

# Filtra las filas con más del 10% de NaN
filtered_surface_temperature = surface_temperature[surface_temperature.apply(lambda row: row.isnull().sum(), axis=1) > threshold]


filtered_surface_temperature.info()
```

```python
# me quedo con las filas que no estan en filtered_surface_temperature
surface_temperature = surface_temperature[~surface_temperature.index.isin(filtered_surface_temperature.index)]
```
