### Distribución por sexos de los refugiados

```python
female = demographics['Female total'].sum()
male = demographics['Male total'].sum()

# Crear un gráfico de barras para mostrar la población total por género
plt.figure(figsize=(8, 6))
bars = plt.bar(['Male', 'Female'], [male, female])

# Agregar el número en cada barra
plt.title('Población Total por Género en el Año 2022')
plt.xlabel('Género')
plt.ylabel('Población Total')
for bar, population in zip(bars, [male, female]):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 100000, f'{population:,}', ha='center', color='black')

plt.show()
```

### Veamos la distrubicón por edad de las mujeres

```python
# quiero haer lo mismo pero por cada grupo de edad y sexo
# Crear un gráfico de barras para mostrar la población total por género

female_teenager = demographics['Female 0 - 4'].sum() + demographics['Female 5 - 11'].sum() + demographics['Female 12 - 17'].sum()
female59 = demographics['Female 18 - 59'].sum()
female60 = demographics['Female 60'].sum()

female_values = [female_teenager, female59, female60]
plt.figure(figsize=(8, 6))
bars = plt.bar(['Female 0-17', 'Female 18-59', 'Female 60+'], female_values)
plt.xticks(rotation=45, ha='right')
```

```python
plt.figure(figsize=(8, 8))
plt.pie(female_values, labels=['Female 0-17', 'Female 18-59', 'Female 60+'], autopct='%1.1f%%', startangle=90)
plt.title('Distribución de Mujeres por Grupos de Edad')
plt.show()
```

### Veamos la distrubicón por edad de los hombres

```python
# quiero haer lo mismo pero por cada grupo de edad y sexo
# Crear un gráfico de barras para mostrar la población total por género

Male_teenager = demographics['Male 0 - 4'].sum() + demographics['Male 5 - 11'].sum() + demographics['Male 12 - 17'].sum()
Male59 = demographics['Male 18 - 59'].sum()
Male60 = demographics['Male 60'].sum()

Male_values = [Male_teenager, Male59, Male60]
plt.figure(figsize=(8, 6))
bars = plt.bar(['Male 0-17', 'Male 18-59', 'Male 60+'], Male_values)
plt.xticks(rotation=45, ha='right')
```

```python
plt.figure(figsize=(8, 8))
plt.pie(Male_values, labels=['Male 0-17', 'Male 18-59', 'Male 60+'], autopct='%1.1f%%', startangle=90)
plt.title('Distribución de Hombres por Grupos de Edad')
plt.show()
```

```python
demographics.groupby('Year')['Total'].sum().plot()
plt.title('Número de refugiados por año')
plt.xlabel('Año')
plt.ylabel('Población')
plt.show()
```

```python
# Suponiendo que 'df_pivot' es un DataFrame pivotado con los años en las columnas y los países en las filas
top_20_countries = demographics.groupby('Country of origin')['Total'].sum().nlargest(20).index
df_pivot = demographics[demographics['Country of origin'].isin(top_20_countries)].pivot_table(values='Total', index='Country of origin', columns='Year', fill_value=0)
sns.heatmap(df_pivot, cmap='magma')
plt.title('Mapa de Calor de Población (Top 20 países)')
plt.xlabel('Año')
plt.ylabel('País')
plt.show()
```
