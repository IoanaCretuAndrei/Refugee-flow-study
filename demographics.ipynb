### DistribuciÃ³n por sexos de los refugiados

```python
female = demographics[demographics['Year']==2022]['Female total'].sum()
male = demographics[demographics['Year']==2022]['Male total'].sum()

# Crear un grÃ¡fico de barras para mostrar la poblaciÃ³n total por gÃ©nero
plt.figure(figsize=(8, 6))
bars = plt.bar(['Male', 'Female'], [male, female])

# Agregar el nÃºmero en cada barra
plt.title('PoblaciÃ³n Total por GÃ©nero en el AÃ±o 2022')
plt.xlabel('GÃ©nero')
plt.ylabel('PoblaciÃ³n Total')
for bar, population in zip(bars, [male, female]):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 100000, f'{population:,}', ha='center', color='black')

plt.show()
```

```python
def plot_demographics(country, year):
    """
    Vemos la composiciÃ³n demogrÃ¡fica de un paÃ­s concreto y su evoluciÃ³n a travÃ©s del tiempo.

    Args:
    country (str):Nombre del paÃ­s.
    year (int): AÃ±o a estudiar.

    Returns:
    None
    """
    # Filtramos por paÃ­s y aÃ±o
    demographics_country_year = demographics[(demographics['Country of origin'] == country) & (demographics['Year'] == year)]

    # Crear una figura con ejes de manera explÃ­cita
    fig = plt.figure(figsize=(50, 30))

    # GrÃ¡fico de barras para la poblaciÃ³n total por gÃ©nero
    female_total = demographics_country_year['Female total'].sum()
    male_total = demographics_country_year['Male total'].sum()

   

    ax1 = fig.add_subplot(221)
    bars_gender = ax1.bar(['Male', 'Female'], [male_total, female_total])
    ax1.set_title(f'Forced Displaced Population in {country} in {year}', fontsize=24)
    ax1.set_xlabel('Gender', fontsize=18)
    ax1.set_ylabel('Total Population', fontsize=18)
    ax1.tick_params(axis='both', labelsize=24)
    ax1.set_ylim(0, 7000000)

    for bar, population in zip(bars_gender, [male_total, female_total]):
        ax1.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 100000, f'{population:,}', ha='center', color='black', fontsize=24)

    # GrÃ¡fico circular para la distribuciÃ³n de edad de las mujeres
    female_age_distribution = [demographics_country_year[f'Female {i}'].sum() for i in ['0 - 4', '5 - 11', '12 - 17', '18 - 59', '60']]
    labels_female = ['Female 0-4', 'Female 5-11', 'Female 12-17', 'Female 18-59', 'Female 60']

    ax4 = fig.add_subplot(224)
    ax4.pie(female_age_distribution, labels=labels_female, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 24})
    ax4.set_title('Distribution of Females by Age Groups', fontsize=24)
    

    # GrÃ¡fico circular para la distribuciÃ³n de edad de los hombres
    male_age_distribution = [demographics_country_year[f'Male {i}'].sum() for i in ['0 - 4', '5 - 11', '12 - 17', '18 - 59', '60']]
    labels_male = ['Male 0-4', 'Male 5-11', 'Male 12-17', 'Male 18-59', 'Male 60']

    ax3 = fig.add_subplot(223)
    ax3.pie(male_age_distribution, labels=labels_male, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 24})
    ax3.set_title('Distribution of Males by Age Groups', fontsize=24)

    # GrÃ¡fico de lÃ­nea para el total de refugiados a lo largo de los aÃ±os
    ax2 = fig.add_subplot(222)
    demographics[demographics['Country of origin'] == country].groupby('Year')['Total'].sum().plot(ax=ax2)
    ax2.set_title('Number of Refugees per Year', fontsize=24)
    ax2.set_xlabel('Year', fontsize=18)
    ax2.set_ylabel('Population', fontsize=18)
    ax2.tick_params(axis='both', labelsize=24)
    ax2.set_ylim(0, 13000000)


    plt.tight_layout()
    plt.show()



```

```python
plot_demographics('Ukraine', 2022)
```
