```python
import matplotlib.pyplot as plt 
from wordcloud import WordCloud 
import pandas as pd
import numpy as np
from PIL import Image
```

```python
df = pd.read_csv(r'C:\Users\ioana\OneDrive\Escritorio\Entorno10\Refugee_project_study\Refugees-flow-study-project\Data\asylum_decisions.csv')


```

```python
#agrupar por pais, por a√±o, y sumar decisiones totales
df1 = df[df['Year']==2022].groupby(['Country of origin'])[ 'Total decisions'].sum()
#eliminar unknown del index
df1 = df1[df1.index.str.strip() != 'Unknown']

```

```python
mask = np.array(Image.open(r'C:\Users\ioana\OneDrive\Escritorio\Entorno10\Refugee_project_study\Refugees-flow-study-project\Data\amorfo.png'))

wc = WordCloud(width=600, height=1000, mask = mask , contour_color='black', contour_width=0.0, background_color='white', colormap='gist_heat_r', max_font_size=150, min_font_size=1 )
wc.generate_from_frequencies(frequencies=df1)
plt.figure( figsize=(20,10) )
plt.axis('off')
plt.imshow(wc)

```
